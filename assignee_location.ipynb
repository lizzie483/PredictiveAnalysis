```python
import pandas as pd
import numpy as np

#Reading the dataset
#Patents of class 424 by year
df = pd.read_csv('assignee_details.csv',low_memory=False)
```

```python
df['assignee_country'].unique()
```

```python
mask = (df['assignee_country']!='None')
df_country = df[mask]
countries = df_country['assignee_country'].unique()
df_assignee_country = pd.DataFrame()
for each_country in countries:
    temp = df_country.loc[(df_country.assignee_country==each_country)]
    total_assignee_by_country = temp['assignee_id'].unique().size
    #print(each_assignee,":",total_patents_by_assignee)
    df_assignee_country = df_assignee_country.append({'assignee_country' : each_country,
                                          'TAC':total_assignee_by_country
                                             },ignore_index=True)
```

```python
df_assignee_country.to_csv('assignee_by_country.csv')
```

```python
#Cleansing data
mask = (df['assignee_id']!='None')
df=df[mask]

assignee = df['assignee_id'].unique()
```

```python
assignee.size
```

```python
#Generating two indicators of prestigio

df_output = pd.DataFrame()
k=0
for each_assignee in assignee:
    k=k+1
    temp = df.loc[(df.assignee_id==each_assignee)]
    
    #get total patents per assignee
    total_patents_by_assignee = temp['patent_number'].unique().size
    print(k,":",each_assignee,":",total_patents_by_assignee)
    df_output = df_output.append({'assignee_id': each_assignee,
                                          'TP' : total_patents_by_assignee
                                         
                                             },ignore_index=True)
```
